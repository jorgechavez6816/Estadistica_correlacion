```python
#Correlaci칩n Pearson con pandas.
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('C:\\Users\\Intel\\Documents\\Mis documentos IDEA\\Samples\\Exportaciones.ILB\\Tabladin치mica10.xlsx')
df.head()
del df['NUM_FACT']
df.head()
df.corr(method='pearson')

#plt.plot(df['COD_PROD_06'], df['TOTAL'], "ro")
#plt.xlabel('Producto 06')
#plt.ylabel('Ventas')
```

```python
#Correlaci칩n Pearson con numpy
import numpy as np

CORR6 = np.corrcoef(df['COD_PROD_06'],df['TOTAL'])
CORR5 = np.corrcoef(df['COD_PROD_05'],df['TOTAL'])
CORR4 = np.corrcoef(df['COD_PROD_04'],df['TOTAL'])
CORR3 = np.corrcoef(df['COD_PROD_03'],df['TOTAL'])
CORR2 = np.corrcoef(df['COD_PROD_02'],df['TOTAL'])
CORR1 = np.corrcoef(df['COD_PROD_01'],df['TOTAL'])

print(CORR6)
print(CORR5)
print(CORR4)
print(CORR3)
print(CORR2)
print(CORR1)
```

```python
#Correlaci칩n Pearson con scipy
from scipy import stats

c6 = stats.pearsonr(df['COD_PROD_06'],df['TOTAL'])
c5 = stats.pearsonr(df['COD_PROD_05'],df['TOTAL'])
c4 = stats.pearsonr(df['COD_PROD_04'],df['TOTAL'])
c3 = stats.pearsonr(df['COD_PROD_03'],df['TOTAL'])
c2 = stats.pearsonr(df['COD_PROD_02'],df['TOTAL'])
c1 = stats.pearsonr(df['COD_PROD_01'],df['TOTAL'])
print(c6)
print(c5)
print(c4)
print(c3)
print(c2)
print(c1)
```

```python
#Correlaci칩n Spearman con pandas
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('C:\\Users\\Intel\\Documents\\Mis documentos IDEA\\Samples\\Exportaciones.ILB\\Tabladin치mica10.xlsx')
df.head()
del df['NUM_FACT']
df.head()
df.corr(method='spearman')
```

```python
#Correlaci칩n Spearman con scipy
from scipy import stats

C6 = stats.spearmanr(df['COD_PROD_06'],df['TOTAL'])
C5 = stats.spearmanr(df['COD_PROD_05'],df['TOTAL'])
C4 = stats.spearmanr(df['COD_PROD_04'],df['TOTAL'])
C3 = stats.spearmanr(df['COD_PROD_03'],df['TOTAL'])
C2 = stats.spearmanr(df['COD_PROD_02'],df['TOTAL'])
C1 = stats.spearmanr(df['COD_PROD_01'],df['TOTAL'])
print(C6)
print(C5)
print(C4)
print(C3)
print(C2)
print(C1)
```

```python
#Correlaci칩n Kendall con scipy
import scipy.stats
from scipy import stats

C6 = scipy.stats.kendalltau(df['COD_PROD_06'],df['TOTAL'])
C5 = scipy.stats.kendalltau(df['COD_PROD_05'],df['TOTAL'])
C4 = scipy.stats.kendalltau(df['COD_PROD_04'],df['TOTAL'])
C3 = scipy.stats.kendalltau(df['COD_PROD_03'],df['TOTAL'])
C2 = scipy.stats.kendalltau(df['COD_PROD_02'],df['TOTAL'])
C1 = scipy.stats.kendalltau(df['COD_PROD_01'],df['TOTAL'])

print(C6)
print(C5)
print(C4)
print(C3)
print(C2)
print(C1)
```
